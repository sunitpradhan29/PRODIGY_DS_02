```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cleanning the data set

```python
df1 = pd.read_csv(r"C:\Users\sunit\prodigy_tasks\titanic\test.csv", encoding='utf-8')
df1
```

```python
df2=pd.read_csv(r"C:\Users\sunit\prodigy_tasks\titanic\train.csv",encoding='utf-8')
df2
```

```python
concat_df = pd.concat([df1, df2], axis=0)
concat_df
```

```python
concat_df.describe()
```

```python
concat_df.drop_duplicates(inplace=True)  
concat_df
```

```python
concat_df.isnull().sum()
```

```python
concat_df['Age'].dtype
avg_age=concat_df['Age'].mean()
avg_age
```

```python
concat_df['Age'] = concat_df['Age'].fillna(avg_age)
```

```python
concat_df['Fare']=concat_df['Fare'].fillna(concat_df['Fare'].mean())
```

```python
concat_df.isnull().sum()
```

```python
concat_df['Cabin'].dtype
```

```python
concat_df=concat_df.drop('Cabin', axis=1)
```

```python
print(concat_df['Embarked'].dtype)
concat_df['Embarked']=concat_df['Embarked'].fillna(concat_df['Embarked'].mode()[0])
```

```python
concat_df['Survived'].dtype
concat_df['Survived']=concat_df['Survived'].fillna(concat_df['Survived'].mean())
```

```python
concat_df.isnull().sum()
```

```python
concat_df.head()
```

```python
concat_df.tail()
```

```python
numerical_columns = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
```

age distribution

```python
plt.figure(figsize=(6,3))
sns.histplot(concat_df['Age'],kde=True,bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=concat_df, x="Sex", hue="Survived")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=concat_df,x='Age',y='Fare',hue='Survived')
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

correlation heat map between age and fare

```python
correlation_matrix = concat_df[['Age', 'Fare']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Heatmap between Age and Fare')
plt.show()
```

correlation heatmap

```python
correlation_matrix = concat_df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
def plot_survival_rate(df):
    fig,ax=plt.subplots()
    ax.pie(df['Survived'].value_counts(), labels=df['Survived'].value_counts().index, autopct='%1.1f%%')
    ax.set_title('Survival Rate')
    plt.show()
plot_survival_rate(concat_df)
```
